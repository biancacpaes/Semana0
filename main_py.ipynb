```python
#letra a)

import random

def obter_escolha_usuario():
    escolha = input("Escolha pedra, papel ou tesoura: ").lower()
    while escolha not in ['pedra', 'papel', 'tesoura']:
        print("Escolha inválida. Tente novamente.")
        escolha = input("Escolha pedra, papel ou tesoura: ").lower()
    return escolha

def obter_escolha_computador():
    return random.choice(['pedra', 'papel', 'tesoura'])

def determinar_vencedor(usuario, computador):
    if usuario == computador:
        return 'empate'
    elif (usuario == 'pedra' and computador == 'tesoura') or \
         (usuario == 'papel' and computador == 'pedra') or \
         (usuario == 'tesoura' and computador == 'papel'):
        return 'usuario'
    else:
        return 'computador'

def jogar_melhor_de_tres():
    pontos_usuario = 0
    pontos_computador = 0

    while pontos_usuario < 2 and pontos_computador < 2:
        usuario = obter_escolha_usuario()
        computador = obter_escolha_computador()
        print(f"\nVocê escolheu: {usuario}")
        print(f"O computador escolheu: {computador}")

        vencedor = determinar_vencedor(usuario, computador)
        if vencedor == 'usuario':
            print("Você ganhou essa rodada!")
            pontos_usuario += 1
        elif vencedor == 'computador':
            print("O computador ganhou essa rodada!")
            pontos_computador += 1
        else:
            print("Empate!")

        print(f"Pontuação: Você {pontos_usuario} x {pontos_computador} Computador\n")

    if pontos_usuario > pontos_computador:
        print("Parabéns, você venceu o jogo!")
    else:
        print("O computador venceu o jogo. Tente novamente!")

jogar_melhor_de_tres()
```
